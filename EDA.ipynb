```python
iris.plot(kind='scatter',x='sepal_length',y='sepal_width',color='r')
plt.show()
```

```python
sn.set_style('whitegrid')
sn.FacetGrid(iris,hue='class',size=4).map(plt.scatter,'sepal_length','sepal_width').add_legend()
plt.show()
```

```python
plt.close()
```

* Pairplot

```python
sn.set_style('whitegrid')
sn.pairplot(iris,hue='class',size=3)
plt.show()
```

* distplot

```python
sn.FacetGrid(iris,hue='class',size=4).map(sn.distplot,'petal_length').add_legend()
plt.show()
```
